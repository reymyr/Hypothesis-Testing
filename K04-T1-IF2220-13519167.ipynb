# 1. Menulis deskripsi statistika (Descriptive Statistics) dari semua kolom pada data yang bersifat numerik, terdiri dari mean, median, modus, standar deviasi, variansi, range, nilai minimum, maksimum, kuartil, IQR, skewness dan kurtosis. Boleh juga ditambahkan deskripsi lain.

Untuk mendapatkan nilai deskripsi statistika dengan python digunakan beberapa fungsi, misal df adalah sebuah variabel dataFrame, maka
- Mean didapatkan dengan fungsi df.mean()
- Median didapatkan dengan fungsi df.median()
- Modus didapatkan dengan fungsi df.mode()
- Standar deviasi didapatkan dengan fungsi df.std()
- Variansi didapatkan dengan fungsi df.var()
- Range didapatkan dengan mengurangi nilai minimum pada nilai maksimum dengan fungsi df.max()-df.min()
- Nilai minimum didapatkan dengan fungsi df.min()
- Nilai maksimum didapatkan dengan fungsi df.max()
- Kuartil didapatkan dengan fungsi df.quantile(x) dengan x menandakan quantile ke-berapa (Contoh x=0.25 untuk kuartil 1)
- IQR didapatkan dengan mengurangi kuartil 1 pada kuartil 3 dengan fungsi df.quantile(0.75)-df.quantile(0.25)
- Skewness didapatkan dengan fungsi df.skew()
- Kurtosis didapatkan dengan fungsi df.kurt(), yaitu definisi kurtosis Fisher

```python
def getStatisticDescription(df):
    print("Mean \t\t=", df.mean())
    print("Median \t\t=", df.median())
    modus = df.mode()
    print("Modus \t\t= ", end="")
    for i in range(len(modus)):
        if i != 0:
            print(", ", end="")
        print(modus[i], end="")
    print()
    print("Standar Deviasi =", df.std())
    print("Variansi \t=", df.var())
    print("Range \t\t=", df.max()-df.min())
    print("Nilai Minimum \t=", df.min())
    print("Nilai Maksimum \t=", df.max())
    print("Q1 \t\t=", df.quantile(0.25))
    print("Q2 \t\t=", df.quantile(0.5))
    print("Q3 \t\t=", df.quantile(0.75))
    print("IQR \t\t=", df.quantile(0.75)-df.quantile(0.25))
    print("Skewness \t=", df.skew())
    print("Kurtosis \t=", df.kurt())
```

# 3. Menentukan setiap kolom numerik berdistribusi normal atau tidak. Gunakan normality test yang dikaitkan dengan histogram plot
Normality test digunakan dengan memanfaatkan library SciPy dengan fungsi stats.normaltest(). Fungsi ini akan menghitung p-value dengan H0 adalah data berdistribusi normal. P-value tersebut lalu dibandingkan dengan taraf signifikansi yang sudah ditentukan dalam hal ini 0.05. Jika nilai p-value yang didapat lebih kecil dari alpha, H0 ditolak atau data tidak berdistribusi normal. Jika tidak, H0 diterima yang berarti data berdistribusi normal. Setelah itu histogram data digambar dan dibandingkan dengan grafik distribusi normal yang memiliki nilai minimal, nilai maksimal, mean, dan standar deviasi yang sama dengan data.

```python
def normalTest(df, alpha):
    k2, p = stats.normaltest(df)
    print("p = {:g}".format(p))
    if p < alpha:  # H0 berarti data berdistribusi normal
        print("H0 ditolak, data tidak berdistribusi normal")
    else:
        print("H0 diterima, data berdistribusi normal")
    
    # Gambar perbandingan histogram dengan kurva distribusi normal
    mu, std = stats.norm.fit(df)
    plt.hist(df, density=True)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    y = stats.norm.pdf(x, mu, std)
    plt.plot(x, y, 'r', lw="3")
    plt.show()
```

# 4. Melakukan test hipotesis 1 sampel, dengan menuliskan 6 langkah testing dan menampilkan juga boxplotnya untuk kolom/bagian yang bersesuaian.

### 6 langkah testing:
1. Tentukan Hipotesis nol (H0 : Î¸ = Î¸0), dimana Î¸ bisa berupa Î¼, Ïƒ^2, p, atau data lain berdistribusi tertentu (normal, binomial, dsc.).

2. Pilih hipotesis alternatif H1 salah dari dari Î¸ > Î¸0, Î¸ < Î¸0, atau Î¸ â‰  Î¸0.

3. Tentukan tingkat signifikan Î±.

4. Tentukan uji statistik yang sesuai dan tentukan daerah kritis.

5. Hitung nilai uji statistik dari data sample. Hitung p-value sesuai dengan uji statistik yang digunakan.

6. Ambil keputusan dengan TOLAK H0 jika nilai uji terletak di daerah kritis atau dengan tes signifikan, TOLAK H0 jika p-value lebih kecil dibanding tingkat signifikansi Î± yang diinginkan

```python
def meanTestZ(xbar, miu0, sigma, n):
    return (xbar - miu0)*math.sqrt(n)/sigma

def proportionTestZ(p, p0, n):
    return (p-p0)/math.sqrt(p0*(1-p0)/n)
```

## 4a. Nilai rata-rata Daerah di atas 4700
- Ditentukan hipotesis nol H0 : Î¼ = 4700
- Kemudian dipilih hipotesis alternatif H1 : Î¼ > 4700
- Selanjutnya ditentukan tingkat signifikan Î± = 0.05
- Akan digunakan uji statistik tes mean menggunakan Z, dan daerah kritis : z > 1.645 diketahui dari P(Z < z) = 0.95
- Akan dihitung nilai uji statistik atau p-valuenya

```python
z = meanTestZ(df["Daerah"].mean(),4700,df["Daerah"].std(),df["Daerah"].count())
print("Nilai uji =", z)
pValue = (1 - stats.norm.cdf(z))
print("P-value =", pValue)

if z > 1.645:
    print("H0 ditolak karena nilai uji terletak di daerah kritis")
else:
    print("H0 diterima karena nilai uji terletak di luar daerah kritis")
if pValue < alpha:
    print("H0 ditolak dikarenakan nilai p-value lebih kecil dibandingkan alpha")
else:
    print("H0 diterima dikarenakan nilai p-value lebih besar sama dengan alpha")

sns.boxplot(data=df["Daerah"], orient="x", width=0.15)
```

### Kesimpulan
Dikarenakan Hipotesis nol ditolak, maka kesimmpulan yang dapat ditarik adalah rata-rata daerah diatas 4700. Hal ini diperkuat dengan nilai p-value yang lebih kecil daripada nilai alpha

## 4b. Nilai rata-rata Sumbu Utama tidak sama dengan 116
- Ditentukan hipotesis nol H0 : Î¼ = 116
- Kemudian dipilih hipotesis alternatif H1 : Î¼ â‰  116
- Selanjutnya ditentukan tingkat signifikan Î± = 0.05
- Akan digunakan uji statistik tes mean menggunakan Z, dan daerah kritis : z < -1.96 dan z > 1.96 diketahui dari P(Z < z) = +-0.975
- Akan dihitung nilai uji statistik atau p-valuenya

```python
z = meanTestZ(df["SumbuUtama"].mean(),116,df["SumbuUtama"].std(),df["SumbuUtama"].count())
print("Nilai uji =", z)
pValue = 2*(1 - stats.norm.cdf(z))
print("P-value =", pValue)

if z < -1.96 or z > 1.96:
    print("H0 ditolak karena nilai uji terletak di daerah kritis")
else:
    print("H0 diterima karena nilai uji terletak di luar daerah kritis")
if pValue < alpha:
    print("H0 ditolak dikarenakan nilai p-value lebih kecil dibandingkan alpha")
else:
    print("H0 diterima dikarenakan nilai p-value lebih besar sama dengan alpha")

sns.boxplot(data=df["SumbuUtama"], orient="x", width=0.15)
```

### Kesimpulan
Dikarenakan Hipotesis nol diterima, maka kesimpulan yang dapat ditarik adalah rata-rata Sumbu Utama adalah 116. Hal ini dibuktikan dengan nilai p-value yang lebih besar daripada nilai alpha

## 4d. Proporsi nilai Diameter yang lebih dari 85, adalah tidak sama dengan 15%
- Ditentukan hipotesis nol H0 : p0 = 0.15
- Kemudian dipilih hipotesis alternatif H1 : p â‰  0.15
- Selanjutnya ditentukan tingkat signifikan Î± = 0.05
- Akan digunakan uji statistik tes proporsi menggunakan Z, dan daerah kritis : z < -1.96 dan z > 1.96 diketahui dari P(Z < z) = 1 - (0.05/2) = 0.975
- Akan dihitung nilai uji statistik atau p-valuenya

```python
pAccent = 0
for i in range(len(df)):
    if df["Diameter"].iloc[i] > 85:
        pAccent += 1
pAccent = pAccent / len(df["Diameter"])

z = proportionTestZ(pAccent, 0.15, len(df["Diameter"]))
print("Nilai uji =", z)
pValue = 2*(1 - stats.norm.cdf(z))
print("P-value =", pValue)

if pValue < alpha:
    print("H0 ditolak dikarenakan nilai p-value lebih kecil dibandingkan alpha")
else:
    print("H1 diterima dikarenakan nilai p-value lebih besar sama dengan alpha")

sns.boxplot(data=df["Diameter"], orient="x", width=0.15)
```

### Kesimpulan
Dikarenakan Hipotesis nol ditolak, maka kesimpulan yang dapat ditarik adalah proporsi data Diameter yang lebih dari 85 bukanlah sebesar 15%.

## 4e. Proporsi nilai Keliling yang kurang dari 100, adalah kurang dari 5%

```python
pAccent = 0
for i in range(len(df)):
    if df["Keliling"].iloc[i] < 100:
        pAccent += 1
pAccent = pAccent / len(df["Keliling"])

z = proportionTestZ(pAccent, 0.05, len(df["Keliling"]))
pValue = stats.norm.cdf(z)
print("Nilai uji =", z)
print("P-value =", pValue)

if pValue < alpha:
    print("H0 ditolak dikarenakan nilai p-value lebih kecil dibandingkan alpha")
else:
    print("H1 diterima dikarenakan nilai p-value lebih besar sama dengan alpha")

sns.boxplot(data=df["Keliling"], orient="x", width=0.15)
```

# 5. Melakukan test hipotesis 2 sampel, dengan menuliskan 6 langkah testing dan menampilkan juga boxplotnya untuk kolom/bagian yang bersesuaian.

### 6 langkah testing:
1. Tentukan Hipotesis nol (H0 : Î¸ = Î¸0), dimana Î¸ bisa berupa Î¼, Ïƒ^2, p, atau data lain berdistribusi tertentu (normal, binomial, dsc.).

2. Pilih hipotesis alternatif H1 salah dari dari Î¸ > Î¸0, Î¸ < Î¸0, atau Î¸ â‰  Î¸0.

3. Tentukan tingkat signifikan Î±.

4. Tentukan uji statistik yang sesuai dan tentukan daerah kritis.

5. Hitung nilai uji statistik dari data sample. Hitung p-value sesuai dengan uji statistik yang digunakan.

6. Ambil keputusan dengan TOLAK H0 jika nilai uji terletak di daerah kritis atau dengan tes signifikan, TOLAK H0 jika p-value lebih kecil dibanding tingkat signifikansi Î± yang diinginkan

```python
def twoSampleMeanTestZ(df1, df2, d0, population):
    return ((df1.mean() - df2.mean()) - d0) / math.sqrt(population.var()/len(df1) + population.var()/len(df2))
```

## 5a. Data kolom AreaBulatan dibagi 2 sama rata: bagian awal dan bagian akhir kolom. Benarkah rata-rata kedua bagian tersebut sama?
Misal Î¼1 adalah rata-rata bagian awal dan Î¼2 adalah rata-rata bagian kedua. Langkah testing:
1. Ditentukan hipotesis nol H0 : Î¼1 - Î¼2 = 0
2. Kemudian dipilih hipotesis alternatif H1 : Î¼1 - Î¼2 â‰  0
3. Selanjutnya ditentukan tingkat signifikan Î± = 0.05
4. Akan digunakan uji statistik tes mean menggunakan Z, dan daerah kritis : z < -1.96 atau z > 1.96 diketahui dari P(Z < z) = 1 - (0.05/2) = 0.975
5. Akan dihitung nilai uji statistik dan p-valuenya

```python
z = twoSampleMeanTestZ(df["AreaBulatan"].head(250), df["AreaBulatan"].tail(250), 0, df["AreaBulatan"])
print("Nilai uji =", z)
pValue = 2 * min(stats.norm.cdf(z), 1-stats.norm.cdf(z))
print("P-value =", pValue)

# Langkah 6. Ambil keputusan
if z < -1.96 or z > 1.96:
    print("H0 ditolak karena nilai uji terletak di daerah kritis")
else:
    print("H0 diterima karena nilai uji terletak di luar daerah kritis")

sns.boxplot(data=[df["AreaBulatan"].head(250), df["AreaBulatan"].tail(250)], orient="x", width=0.15).set_title("Boxplot AreaBulatan1 dan AreaBulatan2")
```

### Kesimpulan
Dikarenakan Hipotesis nol ditolak, maka kesimpulan yang dapat ditarik adalah rata-rata bagian awal dan bagian akhir kolom AreaBulatan berbeda.

## 5d. Proporsi nilai bagian awal Ransum yang lebih dari 2, adalah lebih besar daripada, proporsi nilai yang sama di bagian akhir Ransum?
Misal p1 adalah proporsi bagian awal Ransum yang lebih dari 2 dan p2 adalah proporsi bagian akhir Ransum yang lebih dari 2. Langkah testing:
1. Ditentukan hipotesis nol H0 : p1 - p2 = 0
2. Kemudian dipilih hipotesis alternatif H1 : p1 - p2 > 0
3. Selanjutnya ditentukan tingkat signifikan Î± = 0.05
4. Akan digunakan uji statistik tes proporsi dua sampel menggunakan Z, dan daerah kritis : z > 1.645 diketahui dari P(Z < z) = 0.95
5. Akan dihitung nilai uji statistik dan p-valuenya

```python
# Komputasi
p1 = len(df.head(250)[df.head(250)["Ransum"] > 2]) / 250
p2 = len(df.tail(250)[df.tail(250)["Ransum"] > 2]) / 250
pHat = (len(df.head(250)[df.head(250)["Ransum"] > 2]) + len(df.tail(250)[df.tail(250)["Ransum"] > 2])) / 500

z = (p1 - p2) / (math.sqrt(pHat * (1 - pHat) * (1/250 + 1/250)))
print("Nilai uji =", z)
pValue = 1 - stats.norm.cdf(z)
print("P-value =", pValue)

# Langkah 6. Ambil keputusan
if z > 1.645:
    print("H0 ditolak karena nilai uji terletak di daerah kritis")
else:
    print("H0 diterima karena nilai uji terletak di luar daerah kritis")

sns.boxplot(data=[df["Ransum"].head(250), df["Ransum"].tail(250)], orient="x", width=0.15).set_title("Boxplot Ransum1 dan Ransum2")
```
